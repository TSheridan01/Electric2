```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('Resources/income_all_yrs_ev_count.csv'))
df.head()
```

```python
plt.scatter(df.income_yr_2021, df.ev_count)
plt.scatter(df.income_yr_2020, df.ev_count)
plt.scatter(df.income_yr_2019, df.ev_count)
plt.xlabel('Income in thousands of US dollars')
plt.ylabel('EV count')
plt.title('EV Adoption based on income per county in years 2019, 2020 and 2021')
plt.show()
```

```python
X = df.income_yr_2021.values.reshape(-1, 1)
```

```python
X[:5]
```

```python
y = df.ev_count
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Income in thousands of US dollars')
plt.ylabel('EV count')
plt.title('EV Adoption based on 2021 income per county')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
X = df.income_yr_2020.values.reshape(-1, 1)
model.fit(X, y)
y_pred = model.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Income in thousands of US dollars')
plt.ylabel('EV count')
plt.title('EV Adoption based on 2020 income per county')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
X = df.income_yr_2019.values.reshape(-1, 1)
model.fit(X, y)
y_pred = model.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Income in thousands of US dollars')
plt.ylabel('EV count')
plt.title('EV Adoption based on 2019 income per county')
plt.show()
```
