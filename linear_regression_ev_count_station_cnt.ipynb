```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('Resources/counts_per_county.csv'))
df.head()
```

```python
plt.scatter(df.vehicle_cnt, df.station_cnt)
plt.xlabel('EV count')
plt.ylabel('Fueling stations count')
plt.title('EV Count vs Fueling Stations Count per County')
plt.show()
```

```python
X = df.vehicle_cnt.values.reshape(-1, 1)
```

```python
X[:5]
```

```python
X.shape
```

```python
y = df.station_cnt
```

```python
y.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('EV count')
plt.ylabel('Fueling stations count')
plt.title('EV Count vs Fueling Stations Count per County')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
