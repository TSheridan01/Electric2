```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('Resources/income_perCapita21_station_cnt.csv'))
df
```

```python
plt.scatter(df.per_capita_income_2021, df.stations_count)
plt.xlabel("Per Capita Income")
plt.ylabel('Fueling stations count')
plt.title('Per Capita Houshold Income vs Fueling Stations Count per County')
plt.show()
```

```python
X = df.per_capita_income_2021.values.reshape(-1, 1)
```

```python
X.shape
```

```python
y = df.stations_count
```

```python
y.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel("Per Capita Income")
plt.ylabel('Fueling stations count')
plt.title('Per Capita Houshold Income vs Fueling Stations Count per County')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
model.score(X,y)
```
