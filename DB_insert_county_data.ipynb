```python
# Import our input dataset for county income
us_zip = pd.read_excel('uszips.xlsx')
us_zip = us_zip[['zip', 'lat','lng', 'city','state_id', 'state_name', 'county_fips','county_name']]
us_zip = us_zip.dropna(subset=['zip'])
us_zip.head()

```

```python
#Connecting to database
db_string = f"postgresql://{creds.PGUSER}:{creds.PGPASSWORD}@{creds.PGHOST}:5432/{creds.PGDATABASE}"
engine = create_engine(db_string)
conn = engine.connect()
```

```python
#insert the dataframe to postgress table stripping the index
us_zip.to_sql(name='county', con = conn, if_exists='replace', index=False)
```
