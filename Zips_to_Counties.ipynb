```python
#create function to read data 
def read_data_from_file(name):
    data = []
    with open(name) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')

        # skip the first row
        next(csv_reader)
        for row in csv_reader:
            data.append(row)
    return data
    
```

```python
#create function to create new file 
def NJ_write_features_to_file(name, data, zip_to_county):
       # writing to csv file
    with open(name, 'w', newline='', encoding='utf-8') as csvfile:
        row = ['DMV ID', 'DMV Snapshot(Date)', 'Registration Valid Date', 'ZIP Code', 'VIN Prefix', 'VIN Model Year', 'Registration Expiration Date', 'Make', 'Model', 'Model Year', 'County']
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(row)
        for row in data:
            if str(row[3]) in zip_to_county:
                row.append(zip_to_county[str(row[3])])
            else:
                row.append('-')
            # creating a csv writer object
            csvwriter = csv.writer(csvfile)

            # writing the data rows
            csvwriter.writerow(row)
```

```python
#create function to create new file
def TX_write_features_to_file(name, data, zip_to_county):
    # writing to csv file
    with open(name, 'w', newline='', encoding='utf-8') as csvfile:
        row = ['State', 'ZIP Code', 'Registration Date', 'Vehicle Make', 'Vehicle Model', 'VIN Model Year', 'Drivetrain Type', 'Vehicle GVWR Class', 'Vehicle Category',
               'Vehicle Count', 'DMV Snapshot (Date)', 'DMV Snapshot ID', 'Latest DMV Snapshot Flag', 'County']
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(row)
        for row in data:
            if str(row[1]) in zip_to_county:
                row.append(zip_to_county[str(row[1])])
            else:
                row.append('-')
            # creating a csv writer object
            csvwriter = csv.writer(csvfile)

            # writing the data rows
            csvwriter.writerow(row)
```
