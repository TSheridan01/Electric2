```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('Resources/Capita_income_yr_2021_ev_count.csv'))
df.head()
```

```python
plt.scatter(df.per_capita_income_2021, df.ev_count)
plt.xlabel('Per capita income')
plt.ylabel('EV count')
plt.title('EV Adoption based on per-capita income per county in 2021')
plt.show()
```

```python
X = df.per_capita_income_2021.values.reshape(-1, 1)
```

```python
X[:5]
```

```python
y = df.ev_count
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Per capita income')
plt.ylabel('EV count')
plt.title('EV Adoption based on per-capita income per county in 2021')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
