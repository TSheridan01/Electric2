```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
import sklearn as skl
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
# Import our input dataset
ev_location_df = pd.read_csv('EVStation_income.csv')
ev_location_df.head()
```

```python
ev_location_df.shape
```

```python
# Generate our categorical variable list
ev_location = ev_location_df.dtypes[ev_location_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
ev_location_df[ev_location].nunique()
```

```python
#drop extra statecode of county 

ev_location_df = ev_location_df.drop(columns=['statecode_of_county-2'])
```

```python
ev_location_df.head()
```

```python
# Generate our categorical variable list
ev_location = ev_location_df.dtypes[ev_location_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
ev_location_df[ev_location].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(ev_location_df[ev_location]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names_out(ev_location)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
ev_location_df = ev_location_df.merge(encode_df,left_index=True, right_index=True)
ev_location_df = ev_location_df.drop(ev_location,1)
ev_location_df.head()
```

```python
ev_location_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = ev_location_df["fuel_type_code_ELEC"].values
X = ev_location_df.drop(["fuel_type_code_ELEC"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
